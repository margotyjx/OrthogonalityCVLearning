```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import csv
```

```python
# define collective variables: 2nd and 3rd central moments of coordination numbers
def Dist_Sq_Matrix(x): # matrix of distances squared
    Na = np.size(x,axis = 1)
    D2 = np.zeros((Na,Na))
    for j in range(Na):
        D2[j,:] = (x[0,:]-x[0,j])**2 + (x[1,:]-x[1,j])**2
    return D2

def CV(x):
    Na = np.size(x,axis = 1)
    D2 = Dist_Sq_Matrix(x)
    coord_num = np.zeros((Na,))
    for j in range(Na):
        ind = np.setdiff1d(range(Na),np.array([j]))
        aux = (D2[j,ind]/2.25)**4
        coord_num[j] = np.sum((np.ones((1,Na-1))-aux)/(np.ones((1,Na-1))-aux**2))
    cmean = np.mean(coord_num)
    mu2 = np.mean((coord_num-cmean)**2)
    mu3 = np.mean((coord_num-cmean)**3)
    return mu2,mu3
```

```python
# read data from txt files
import re
# parameters
# beta = 5
# gamma = 0.1
# sigma = 0.02
# fname = "GaussianBumps_beta5_gamma0p1_h1e-3.txt"

# beta = 5
# gamma = 0.05
# sigma = 0.02

# beta = 5
# gamma = 0.5
# sigma = 0.02

# beta = 5
# gamma = 1
# sigma = 0.01

# beta = 5
# gamma = 1
# sigma = 0.02

beta = 5
gamma = 0.5
sigma = 0.02
fname = "GaussianBumps_beta5.txt"

Gbumps = [] 
with open(fname, "r") as f:
    for line in f:
        # cleaning the bad chars in line
        line = line.strip()
        line = line.strip(" \\n")
        line = re.sub(r"(-[0-9]+\.)", r" \1", line)
        values = [float(value) for value in line.split()]
        Gbumps.append(values)
print(np.shape(Gbumps))
Gbumps = np.array(Gbumps)
print(np.shape(Gbumps))
```

```python
# plot centers of Gaussian bumps 
plt.rcParams.update({'font.size': 20})
plt.figure(figsize=(8,8))
plt.scatter(Gbumps[:,1],Gbumps[:,2])
axes=plt.gca()
#axes.set_aspect(1)
axes.set_xlabel('$\mu_2$')
axes.set_ylabel('$\mu_3$')
```

```python
Nbumps,m = np.shape(Gbumps)
print(Nbumps,m)
```

```python
# compute the added potential
def Gbumps_potential(x,y):
    sig = 0.02
    sig2 = sig*sig
    GBpot = np.sum(Gbumps[:,0]*np.exp(-0.5*((x-Gbumps[:,1])**2 + (y-Gbumps[:,2])**2)/sig2))
    return GBpot

Ngrid = 100
mu2min = 0.2
mu2max = 1.4
mu3min = -0.7
mu3max = 1.4
mu2lin = np.linspace(mu2min,mu2max,Ngrid)
mu3lin = np.linspace(mu3min,mu3max,Ngrid)
mu2step = mu2lin[1]-mu2lin[0]
mu3step = mu3lin[1]-mu3lin[0]
mu2grid,mu3grid = np.meshgrid(mu2lin,mu3lin)

VGB = np.zeros((Ngrid,Ngrid))
for i in range(Ngrid):
    for j in range(Ngrid):
        VGB[i,j] = Gbumps_potential(mu2grid[i,j],mu3grid[i,j])


FE = -VGB*gamma/(gamma + 1/beta)   
FEmin = np.min(FE)
print(FEmin)
FE = FE - FEmin
        
plt.rcParams.update({'font.size': 20})
plt.figure(figsize=(8,8))
plt.contourf(mu2grid,mu3grid, FE, levels=15, cmap='turbo')
cbar = plt.colorbar()
cbar.set_label(r'Gaussian Bumps potential ($a^{-1}$)', size=20)
plt.xlim([mu2min,mu2max])
plt.ylim([mu3min,mu3max])
plt.xlabel(r'$\mu_2$')
plt.ylabel(r'$\mu_3$')
axes=plt.gca()
#axes.set_aspect(1)
plt.savefig("FE_LJ7in2Dbeta5.pdf")
```
